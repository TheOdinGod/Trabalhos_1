```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Baixando a base de dados
df = pd.read_csv("/content/ataques_seguranca.csv", sep = ";")
df.head()
```

```python
# Análise Temporal
# # Tempo de resolução de incidentes
# # Analise como o tempo de resolução de incidentes  varia ao longo dos anos

import pandas as pd
import matplotlib.pyplot as plt

# Agrupar dados por ano e calcular o tempo médio de resolução de incidentes
incident_resolution_by_year = df.groupby('Year')['Incident Resolution Time (in Hours)'].mean()

# Criando a lógica
plt.figure(figsize=(10, 6))
plt.plot(incident_resolution_by_year.index, incident_resolution_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Mean Incident Resolution Time (Hours)')
plt.title('Incident Resolution Time Variation Over Years')
plt.grid(True)
plt.show()
```

```python
# Comparação de perdas financeiras ao longo dos anos:
# Analisar a variação da coluna Financial Loss (in Million $) agrupada por Year.

import matplotlib.pyplot as plt
# Agrupar dados por ano e somar perdas financeiras
financial_loss_by_year = df.groupby('Year')['Financial Loss (in Million $)'].sum()

# Criando a lógica
plt.figure(figsize=(10, 6))
plt.plot(financial_loss_by_year.index, financial_loss_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Financial Loss (in Million $)')
plt.title('Financial Loss Variation Over Years')
plt.grid(True)
plt.show()
```

```python
# Setores mais afetados por ataques cibernéticos:
# Comparar a quantidade de ataques por Target Industry para identificar os mais vulneráveis.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que seu DataFrame seja chamado de 'df' e já esteja carregado

# Contagem de ataques por setor-alvo
attack_counts = df['Target Industry'].value_counts()

# Crie um gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_counts.index, y=attack_counts.values)
plt.xlabel('Target Industry')
plt.ylabel('Number of Attacks')
plt.title('Number of Cyber Attacks per Target Industry')
plt.xticks(rotation=45, ha='right')  # gire os rótulos do eixo x para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python
# Tipos de ataque mais comuns:
# Contar a frequência de cada Attack Type e verificar qual é o mais recorrente.

import matplotlib.pyplot as plt
# Conte a frequência de cada tipo de ataque
attack_type_counts = df['Attack Type'].value_counts()

# Imprima a frequência de cada tipo de ataque
print(attack_type_counts)

# Encontre o tipo de ataque mais frequente
most_frequent_attack = attack_type_counts.index[0]
print(f"\nThe most frequent attack type is: {most_frequent_attack}")

# Crie um gráfico de barras de frequências de tipo de ataque
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_type_counts.index, y=attack_type_counts.values)
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.title('Frequency of Different Attack Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Origem dos ataques mais frequentes:
# Analisar a distribuição da coluna Attack Source e identificar as principais fontes de ataque.

import matplotlib.pyplot as plt
# Supondo que seu DataFrame seja chamado de 'df' e já esteja carregado

# Conte a frequência de cada fonte de ataque
attack_source_counts = df['Attack Source'].value_counts()

# Imprima a frequência de cada fonte de ataque
print(attack_source_counts)

# Encontre a fonte de ataque mais frequente
most_frequent_source = attack_source_counts.index[0]
print(f"\nThe most frequent attack source is: {most_frequent_source}")

# Crie um gráfico de barras de frequências de fontes de ataque
plt.figure(figsize=(12, 6))
sns.barplot(x=attack_source_counts.index, y=attack_source_counts.values)
plt.xlabel('Attack Source')
plt.ylabel('Number of Attacks')
plt.title('Frequency of Attacks from Different Sources')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Tempo médio de resolução por tipo de ataque:
# Calcular o tempo médio de Incident Resolution Time (in Hours) para cada Attack Type.

import matplotlib.pyplot as plt
# Calcule o tempo médio de resolução de incidentes para cada tipo de ataque
mean_resolution_time_by_attack_type = df.groupby('Attack Type')['Incident Resolution Time (in Hours)'].mean()

# Imprima os resultados
print(mean_resolution_time_by_attack_type)

# Crie um gráfico de barras para visualizar os resultados
plt.figure(figsize=(12, 6))
sns.barplot(x=mean_resolution_time_by_attack_type.index, y=mean_resolution_time_by_attack_type.values)
plt.xlabel('Attack Type')
plt.ylabel('Mean Incident Resolution Time (Hours)')
plt.title('Mean Incident Resolution Time by Attack Type')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Eficiência dos mecanismos de defesa:
# Comparar a eficiência de cada Defense Mechanism Used em reduzir o Incident Resolution Time (in Hours).

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que seu DataFrame seja chamado de 'df' e já esteja carregado

# Agrupar dados por 'Mecanismo de Defesa Usado' e calcular o tempo médio de resolução de incidentes
resolution_time_by_defense = df.groupby('Defense Mechanism Used')['Incident Resolution Time (in Hours)'].mean()

# Crie um gráfico de barras para visualizar os resultados
plt.figure(figsize=(12, 6))
sns.barplot(x=resolution_time_by_defense.index, y=resolution_time_by_defense.values)
plt.xlabel('Defense Mechanism Used')
plt.ylabel('Mean Incident Resolution Time (Hours)')
plt.title('Mean Incident Resolution Time by Defense Mechanism')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Imprima o tempo médio de resolução de incidentes para cada mecanismo de defesa
resolution_time_by_defense
```

```python
# Países com maiores perdas financeiras:
# Agrupar por Country e somar os valores de Financial Loss (in Million $).

import matplotlib.pyplot as plt
# Agrupar por país e somar perdas financeiras
financial_loss_by_country = df.groupby('Country')['Financial Loss (in Million $)'].sum()

# Classificar por perda financeira em ordem decrescente
financial_loss_by_country = financial_loss_by_country.sort_values(ascending=False)

# Imprima os principais países com as maiores perdas financeiras
print(financial_loss_by_country)

# Crie um gráfico de barras de perdas financeiras por país
plt.figure(figsize=(12, 6))
financial_loss_by_country.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Total Financial Loss (in Million $)')
plt.title('Total Financial Loss by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
