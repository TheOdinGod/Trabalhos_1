```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/ataques_seguranca.csv', sep=';')
df.head()
```

```python
df['Country'].value_counts().head(10).plot(kind='bar', title='Top 10 países com mais ataques')
```

```python
df['Attack Type'].value_counts().plot(kind='bar', title='Tipos de ataque mais comuns')
```

```python
df['Target Industry'].value_counts().plot(kind='barh', title='Setores mais atacados')
```

```python
df['Security Vulnerability Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Tipos de vulnerabilidade')
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(14,10))

# País
df['Country'].value_counts().head(5).plot(kind='bar', ax=axs[0,0], title='Top países com mais ataques')

# Tipo de ataque
df['Attack Type'].value_counts().head(5).plot(kind='bar', ax=axs[0,1], title='Tipos de ataque')

# Setor alvo
df['Target Industry'].value_counts().head(5).plot(kind='barh', ax=axs[1,0], title='Setores mais atacados')

# Vulnerabilidade
df['Security Vulnerability Type'].value_counts().head(5).plot(kind='bar', ax=axs[1,1], title='Vulnerabilidades mais comuns')

plt.tight_layout()
plt.show()
```
