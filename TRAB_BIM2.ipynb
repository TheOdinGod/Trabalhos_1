```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/ataques_seguranca.csv', sep=';')
df.head(3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Gráfico 1 - Países com mais ataques
df['Country'].value_counts().head(5).plot(kind='bar', color='steelblue', figsize=(5,3), title='Países mais atacados')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 2 - Tipos de ataque mais comuns
df['Attack Type'].value_counts().head(5).plot(kind='bar', color='indianred', figsize=(5,3), title='Tipos de ataque')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 3 - Setores mais atacados
df['Target Industry'].value_counts().head(5).plot(kind='barh', color='mediumseagreen', figsize=(5,3), title='Setores mais atacados')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python



# Gráfico 4 - Vulnerabilidades mais comuns
df['Security Vulnerability Type'].value_counts().head(5).plot(kind='bar', color='darkorange', figsize=(5,3), title='Vulnerabilidades mais comuns')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
# Gráfico 1 - Países
df['Country'].value_counts().head(5).plot(kind='bar', ax=axs[0, 0], color='steelblue', title='Países mais atacados')
axs[0, 0].set_xlabel('')
axs[0, 0].set_ylabel('')
# Gráfico 2 - Tipos de ataque
df['Attack Type'].value_counts().head(5).plot(kind='bar', ax=axs[0, 1], color='indianred', title='Tipos de ataque')
axs[0, 1].set_xlabel('')
axs[0, 1].set_ylabel('')
# Gráfico 3 - Setores
df['Target Industry'].value_counts().head(5).plot(kind='barh', ax=axs[1, 0], color='mediumseagreen', title='Setores mais atacados')
axs[1, 0].set_xlabel('')
axs[1, 0].set_ylabel('')
# Gráfico 4 - Vulnerabilidades
df['Security Vulnerability Type'].value_counts().head(5).plot(kind='bar', ax=axs[1, 1], color='darkorange', title='Vulnerabilidades mais comuns')
axs[1, 1].set_xlabel('')
axs[1, 1].set_ylabel('')
plt.tight_layout(pad=1.0)
plt.show()
```
